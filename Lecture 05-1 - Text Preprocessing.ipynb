# 1. Loading data

```python
import nltk
f = open('Data/harry_7books.txt', encoding='utf8')
raw = f.read()
tokens = nltk.word_tokenize(raw)
```

```python
print(tokens[1000:1100])
```

# 2. Regular expression

```python
tokens = [w.lower() for w in tokens] # แปลง w ที่เป็น token ให้เป็นตัวพิมพ์เล็กทั้งหมด
```

```python
import re
[w for w in tokens if re.search('ed$', w)] 

# $ = ลงท้าย
# 'ed$' = ลงท้ายด้วย ed
```

```python
[w for w in tokens if re.search('^..j..t..$', w)] 

# ^ = เริ่มต้น
# .. = 1 character ใด ๆ 
# $ = ลงท้าย
```

```python
[w for w in tokens if re.search('^w.?ed$', w)] 
```

```python
[w for w in tokens if re.search('^w.*ed$', w)] 
```

```python
[w for w in tokens if re.search('^spo.ts?$', w)]  # ? = อะไรก็ตามที่อยู่ข้างหน้า ? จะเป็น optional
```

```python
[w for w in tokens if re.search('^[b-f]', w)] # '^[b-f]' = ขึ้นต้นด้วย b ถึง f
```

```python
[w for w in tokens if re.search('^[bdf][aue][rts]$', w)] 

# กำหนดจำนวน character ได้ 
# โดย '^[bdf][aue][rts]$' = จะขึ้นต้นด้วยตัวอักษร b หรือ d หรือ f ตัวต่อมาจะเป็นตัวอักษร a หรือ u หรือ e และลงท้ายด้วยตัวอักษร r หรือ t หรือ s
```

```python
set([w for w in tokens if re.search('noo+$', w)]) # + = 1 หรือมากกว่านั้น
```

```python
set([w for w in tokens if re.search('noo*$', w)]) # * = 0 หรือมากกว่านั้น
```

```python
[w for w in tokens if re.search('\.+$', w)] 

# \. = จะช่วยในการบอกว่านี่ไม่ใช่ Regular expression นะ 
# '\.+$' = ต้องการให้ Detect จุด(.) 1 อัน หรือมากกว่าขึ้นไปในตัวสุดท้าย
```

```python
[w for w in tokens if re.search('^[0-9]{4}$', w)] # {4} = 4 character 
```

```python
nltk.download('treebank')
wsj = sorted(set(nltk.corpus.treebank.words()))
[w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]

# ^[0-9]+ = ตัวแรกเป็น 0 ถึง 9 มีตั้งแต่ 1 ตัวขึ้นไป
# - = ขีด
#[a-z]{3,5}$ = a ถึง z โดยให้มี 3 ถึง 5 ตัวขึ้นไป
```

```python
[w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)] 

# {5,} = อย่างน้อย 5 ตัว 
# {,6} = อย่างมาก 6 ตัว 
```

```python
set([w for w in tokens if re.search('(ed|ing)$', w)]) 

# | = เลือกอันใดอันหนึ่ง
# () = ชุดนึงสามารถใส่ operater ใด ๆ ไปได้
```

# 2. Regular expression tokenizer

## 2.1 Basic regex tokenizer

```python
s = ("Good muffins cost $3.88\nin New York.  Please buy me\n"
      "two of them.\n\nThanks.")
s2 = ("Alas, it has not rained today. When, do you think, "
      "will it rain again?")
s3 = ("<p>Although this is <b>not</b> the case here, we must "
      "not relax our vigilance!</p>")
```

```python
s2
```

```python
nltk.regexp_tokenize(s2, r'[,\.\?!"]\s*', gaps=True) # gaps=True คือ ให้เอาทุกอย่าง ยกเว้น,\.\?!" มาเป็น token
```

```python
s3
```

```python
nltk.regexp_tokenize(s3, r'</?.>', gaps=False) # gaps=False คือ ให้เอาเฉพาะ pattern นี้ </?.>
```

```python
nltk.regexp_tokenize(s3, r'</?.>', gaps=True)
```

## 2.2 Tokenization and removing all punctuations

```python
tokenizer = nltk.RegexpTokenizer(r"\w+") # \w+ คือ a-zA-Z0-9 1 ตัวขึ้นไป
tokens_0 = tokenizer.tokenize(raw)
print(tokens_0[2000:2100])
```

# 3. Text normalization

## 3.1 Transform to lower cases

```python
tokens = [w.lower() for w in tokens]
tokens
```

## 3.2 Replacing \xad

```python
set([w.replace('\xad','') for w in tokens if w.startswith('gry') ])  
```

```python
tokens_1 = [w.replace('\xad','') for w in tokens] # แทน \xad ด้วยโล่ง ๆ
```

## 3.3 Stemming

    Porter Stemming คือ เอาคำลงท้ายในภาษาอังกฤษ โดยไม่ใช้ Dictionary
    ตัวอย่างการลดรูป: sses --> ss
                   ies --> i
                   ational --> ate
                   tional --> tion

```python
porter = nltk.PorterStemmer()
tokens_2 = [porter.stem(t) for t in tokens_1]
```

```python
tokens_1[1000:1100]
```

```python
tokens_2[1000:1100]
```

    จะเห็นว่าคำในนี้ได้ถูกลดรูปไปแล้ว แต่ในบางคำยังผิด เช่น จาก his กลายเป็น hi

## 3.4 Lemmatization

    Lemmatization คือ ลดรูปคำที่มีความหมายเดียวกัน โดยการ map จากคำชุดนึงไปอีกชุดนึง
    ตัวอย่างการลดรูป: am, are, is --> be
                   car, cars, car's, cars' --> car

```python
nltk.download('wordnet')
wnl = nltk.WordNetLemmatizer()
  
print("rocks :", wnl.lemmatize("rocks")) 
print("corpora :", wnl.lemmatize("corpora")) 
print("running :", wnl.lemmatize("running", pos='v')) # pos='v' คือ part of speech: verb
print("better :", wnl.lemmatize("better ", pos='a')) # pos='a' คือ part of speech: adjective 
```

## 3.5 Remove stopwords

    Stopwords คือ คำที่เกิดขึ้นบ่อย ๆ เช่น พวกคำเชื่อม

### 3.5.1 Stopword list

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
```

### 3.5.2 Stopword removal

```python
print(tokens[2000:2100])
```

```python
tokens_4 = [w for w in tokens if not w in stop_words]
print(tokens_4[2000:2100])
```

## 3.6 POS Tagging

    Part of Speech (POS) Tagging คือ การ Tag ว่าคำเหล่านี้คือประเภทอะไร

```python
nltk.download('averaged_perceptron_tagger')
tokens_5 = nltk.pos_tag(tokens)
```

```python
tokens_5[2000:2100]
```

## Word Frequency Analysis

```python
from collections import Counter

counter = Counter(tokens_4)
print(counter.most_common(10))
```

```python
import pandas as pd
df = pd.DataFrame(tokens_4)
```

```python
import matplotlib.pyplot as plt

df_stack = df.stack(level=0) 
counter = df_stack.value_counts()[0:10] # top 10
plt.barh(counter.index, counter.values, color = '#dd94aa' ) 
plt.gca().invert_yaxis()
plt.title("Top 10 Words in Harry Potter and the Chamber of Secrets")
plt.xlabel('Frequency') 
plt.ylabel('Word')
plt.show()
```
